## Model Datapreprocessing and Preparation For Data

```python
dataset = pd.read_csv('sonardata.csv',header=None)
```

```python
dataset.head()
```

```python
dataset[1]
```

```python
dataset.shape
```

```python
dataset[60].value_counts()
```

```python
dataset.groupby(60).mean()
```

```python
X = dataset.drop(columns = 60 ,axis = 1)
y = dataset[60]
```

```python
print(X)
print(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1 , stratify = y , random_state = 1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

# train the model with Data

```python
model.fit(X_train,y_train)
```

```python
print(y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = classification_report(X_train_prediction,y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
training_data_accuracy_test = classification_report(X_test_prediction,y_test)
```

```python
print(training_data_accuracy_test)
```
